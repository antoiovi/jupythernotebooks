## Convertiamo il file on-line

La conversione del file la facciamo tramite comandi python.
Per nasconde i comandi nel file html esportatao vedere il paragrafo successivo.
Il file si esporta in html sempre con il comando :

``$ jupyter nbconvert --to html ExportImages.ipynb ``

```python
import base64, io, IPython
from PIL import Image as PILImage
image_path="jupyter.jpeg"
image = PILImage.open(image_path)
output = io.BytesIO()
image.save(output, format='PNG')
encoded_string = base64.b64encode(output.getvalue()).decode()
html = '<img src="data:image/png;base64,{}"/>'.format(encoded_string)
IPython.display.HTML(html)
```
