# Appunti sulla covarianza
Vogliamo sperimentare, usando Python, sui legami tra due variabili casuali.
Ovvero vogliamo analizzare i valori di varianza, covarianza, coefficiente di correlazione.
Utilizziamo per creare questi appunti e per eseguire il codice, il blocco notes jupyter python. 

#### Prima di tutto importiamo le librerie di interesse :

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Variabili direttamente (e fortemente) correlate
Creiamo poi due variabili casuali x,e y e stampiamo alcuni dati.
Prima di tutto le due variabili (per semplificare con solo 4 valori) le creiamo fortemente correlate:

```python
x=[5,10,5,10]
y=[4,9,4,9]
print("Covarinace (x,y):\n {}".format(np.cov(x,y)))
print("Correlation coefficient (x,y): \n {}".format(np.corrcoef(x,y)))
print("Variance (x)= {} Variance (y)= {} ".format(np.var(x),np.var(y)))

fig = plt.figure()   
plt.plot(x)
plt.plot(y)
plt.show()
```


Come si vede le due variabili sono fortemente correlate, ed il coefficiente di correlazione è uguale ad 1. 
La matrice di covarianza np.cov(x,y):

[[C11 C12]
[C21 C22]]

restituisce sulla diagonale Cii i valori della covarianza della variabile xi con se stessa
Al valore Cij il valore della covarianza tra xi e xj.
Nel nostro esempio xi è la variabile x, xj è la variabile y.
La matrice di correlazione R= np.corrcoef(x,y):
[[R11 R12]
[R21 R22]]
restituisce sulla diagonale Rii i valori della coRRelazione della variabile xi con se stessa
Al valore Rij il valore della coRRelazione tra xi e xj.
Nel nostro esempio xi è la variabile x, xj è la variabile y.
Variabili inversamente correlate
Modifichiamo e due variabili per renderle inversamente correlate. Quando una aumenta , l'altra diminuisce.



```python

x=[5,10,5,10]
y=[10,5,10,5]
print("Covarinace (x,y):\n {}".format(np.cov(x,y)))
print("Correlation coefficient (x,y): \n {}".format(np.corrcoef(x,y)))
print("Variance (x)= {} Variance (y)= {} ".format(np.var(x),np.var(y)))

fig = plt.figure()   
plt.plot(x)
plt.plot(y)
plt.show()
```

Come si vede , ora la matrice di correlazione presenta due diagonali uguali a -1, e la matrice delle covarianze ha una diagonale negativa.
Le variabili sono INVERSAMENTE correlate.
Variabili debolmente correlate

```python
x=[5,10,5,10]
y=[1,2,3,4]
print("Covarinace (x,y):\n {}".format(np.cov(x,y)))
print("Correlation coefficient (x,y): \n {}".format(np.corrcoef(x,y)))
print("Variance (x)= {} Variance (y)= {} ".format(np.var(x),np.var(y)))

fig = plt.figure()   
plt.plot(x)
plt.plot(y)
plt.show()
```

 Abbiamo un fattore di correlazione uguale a 0.447. 
 Le variabili sono moderatamente correlate in positivo.
