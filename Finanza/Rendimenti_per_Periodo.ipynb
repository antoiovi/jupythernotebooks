````python
#['AIM', 'FTMIBSHARES', 'MTA', 'MTAAUCNOMKTORD', 'MTAInternat.', 'MTANOMKTORD', 'STAR']
f="/home/antoiovi/datiborsa/azioni.csv"
def get_ALL():
    df=pd.read_csv(f,sep=';')
    return df
def get_FTSMIB():
    df=pd.read_csv(f,sep=';')
    return df[df['Mercato']=='FTMIBSHARES']
def get_MTA():
    df=pd.read_csv(f,sep=';')
    return df[df['Mercato']=='MTA']

def get_AIM():
    df=pd.read_csv(f,sep=';')
    return df[df['Mercato']=='AIM']

def get_STAR():
    df=pd.read_csv(f,sep=';')
    return df[df['Mercato']=='AIM']
````

````python
df1=pd.DataFrame()
dfstar=get_STAR()
for k,v in dfstar.iterrows():
    t=v['ticker']
    df = yf.Ticker(t)
    # valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
    # (optional, default is '1d')
    df = df.history(period="4mo",interval='1mo')
    df['delta']=df['Close']-df['Open']
    df['delta_perc']=100*df['delta']/df['Open']
    df1[t]=df['delta_perc']
df1
````

````python
df=pd.DataFrame(df1.unstack())
df.reset_index(inplace=True)
#
df.columns=['sigla','data','rend']
df
````

````python
df.groupby(pd.Grouper(key='data',freq='1M'),pd.Grouper(key='data'))['rend'].mean()
````
