# NOTA : La data in formato utc

Scaricando i dati della borsa italiana si ha l'ora italiana con il +2 o piu 1


```python
from datetime import datetime
from dateutil import tz

# METHOD 1: Hardcode zones:
from_zone = tz.gettz('UTC')
to_zone = tz.gettz('Europe/Rome')

# METHOD 2: Auto-detect zones:
#from_zone = tz.tzutc()
#to_zone = tz.tzlocal()
```

```python
import yfinance as yf
data = yf.download("STM.MI", start="2021-05-14", end="2021-05-15",interval="1h")

data['local']=data.apply(lambda x : x.name.astimezone(to_zone), axis=1)
```
