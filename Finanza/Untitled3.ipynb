```python
import pandas as pd
import numpy as np
from datetime import datetime
from matplotlib import pyplot as plt
#01/10/2020 13:20:00,2.7915,2.7915,2.7915,2.7915,700,3SIL,0
```

```python
names=['data','Open','Close','High','Low','Volume','Symbol','Interp']
mydateparser = lambda x: pd.datetime.strptime(x, "%d/%m/%Y %H:%M:%S")

sil="/home/antoiovi/datiborsa/3sil.csv"
df=pd.read_csv(sil,parse_dates=['data'],date_parser=mydateparser,sep=",",
                 names=names,decimal=',',index_col='data')
df['High']=pd.to_numeric(df['High'])
```

```python

df=df.resample('1D').agg({'High': 'max', 'Volume': 'sum'})
df=df.sort_values('Volume').dropna()
df.head(5)
```

```python
df1=df[df['Volume']>1000000]
df2=df[df['Volume']<1000000]
```
